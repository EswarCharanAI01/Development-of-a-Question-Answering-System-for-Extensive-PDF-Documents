````python
from torch import cuda, bfloat16
import transformers

model_id =  "meta-llama/Llama-2-7b-chat-hf"#'meta-llama/Llama-2-7b-chat-hf'

device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'

# set quantization configuration to load large model with less GPU memory
# this requires the `bitsandbytes` library
bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)

# begin initializing HF items, you need an access token
#hf_auth = '<add your access token here>'
hf_auth = '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'
model_config = transformers.AutoConfig.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)

model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    config=model_config,
    quantization_config=bnb_config,
    device_map='auto',
    use_auth_token=hf_auth
)

# enable evaluation mode to allow model inference
model.eval()

print(f"Model loaded on {device}")
````

````python
tokenizer = transformers.AutoTokenizer.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)
````

````python
stop_list = ['\nHuman:', '\n```\n']

stop_token_ids = [tokenizer(x)['input_ids'] for x in stop_list]
stop_token_ids
````

````python
import torch

stop_token_ids = [torch.LongTensor(x).to(device) for x in stop_token_ids]
stop_token_ids
````

````python
from transformers import StoppingCriteria, StoppingCriteriaList

# define custom stopping criteria object
class StopOnTokens(StoppingCriteria):
    def __call__(self, input_ids: torch.LongTensor, scores: torch.FloatTensor, **kwargs) -> bool:
        for stop_ids in stop_token_ids:
            if torch.eq(input_ids[0][-len(stop_ids):], stop_ids).all():
                return True
        return False

stopping_criteria = StoppingCriteriaList([StopOnTokens()])
````

````python
generate_text = transformers.pipeline(
    model=model,
    tokenizer=tokenizer,
    return_full_text=True,  # langchain expects the full text
    task='text-generation',
    # we pass model parameters here too
    stopping_criteria=stopping_criteria,  # without this model rambles during chat
    temperature=0.1,  # 'randomness' of outputs, 0.0 is the min and 1.0 the max
    max_new_tokens=512,  # max number of tokens to generate in the output
    repetition_penalty=1.1  # without this output begins repeating
)
````

````python
res = generate_text("Explain me the difference between Data Lakehouse and Data Warehouse.")
print(res[0]["generated_text"])
````

````python
from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=generate_text)

# checking again that everything is working fine
#llm(prompt="Explain me the difference between Data Lakehouse and Data Warehouse.")
````

````python
import PyPDF2

# Assuming you have a list of PDF file paths
pdf_files = ["/content/blade runner 2049.pdf"]

def load_pdf(file_path):
    with open(file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]  # Updated method here
            text += page.extract_text()
        return text

# Load each PDF file and store the extracted text
documents = [load_pdf(file) for file in pdf_files]
````

````python
for index, text in enumerate(documents):
    print(f"Extracted text from document {index + 1}:")
    print(text)
````

````python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Assuming 'documents' contains the extracted text from PDF files as strings
# documents = [
#     "Text extracted from PDF file 1...",
#     "Text extracted from PDF file 2...",
#     # Add more extracted text strings here for each PDF file
# ]

# Define chunk size and overlap
chunk_size = 1000
chunk_overlap = 20

# Split the documents into chunks
all_splits = []
for doc in documents:
    chunks = []
    start_idx = 0
    while start_idx < len(doc):
        end_idx = min(start_idx + chunk_size, len(doc))
        chunk = doc[start_idx:end_idx]
        chunks.append(chunk)
        start_idx += chunk_size - chunk_overlap
    all_splits.append(chunks)

# Now 'all_splits' contains a list of lists, where each inner list represents the chunks of text for a document
````

````python
# Print the chunks for each document
for i, chunks in enumerate(all_splits):
    print(f"Chunks for Document {i + 1}:")
    for j, chunk in enumerate(chunks):
        print(f"Chunk {j + 1}: {chunk}")
    print()  # Add an empty line between documents for clarity
````

````python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {"device": "cuda"}

# Assuming 'documents' contains the extracted text from PDF files as strings
# documents = [
#     "Text extracted from PDF file 1...",
#     "Text extracted from PDF file 2...",
#     # Add more extracted text strings here for each PDF file
# ]

# Create an instance of HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name=model_name, model_kwargs=model_kwargs)

# Convert strings to Document objects with metadata
class Document:
    def __init__(self, text):
        self.page_content = text  # Assign the text to page_content
        self.metadata = {}  # Empty metadata for simplicity

# Convert strings to Document objects
documents = [Document(text) for text in documents]

# Storing embeddings in the vector store
vectorstore = FAISS.from_documents(documents, embeddings)
````

````python
from langchain.chains import ConversationalRetrievalChain

chain = ConversationalRetrievalChain.from_llm(llm, vectorstore.as_retriever(), return_source_documents=True)
````

````python
chat_history = []

query = "Explain the theme of the movie?"
result = chain({"question": query, "chat_history": chat_history})

print(result['answer'])
````
